## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?
table =orders.groupby(['CustomerID']).agg({'amount_spent':sum}).reset_index()
table
```

```python
#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?
#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?
def customer_priority(x):
    if x>np.percentile(table['amount_spent'],95):
        return 'VIP'
    elif x>np.percentile(table['amount_spent'],75) and x<=np.percentile(table['amount_spent'],95):
        return 'Preferred'
    else:
        return 'Unknown'

table['customer_priority']=table['amount_spent'].apply(customer_priority)
table.head(10)
```

## Q2: How to identify which country has the most VIP Customers?

```python
final_table = pd.merge(left = orders,
                       right = table, 
                       how = 'inner', 
                       left_on = "CustomerID", 
                       right_on= "CustomerID")

# df_orders_cust_tiers = df_orders_cust_tiers.rename(columns={'amount_spent_x': 'amount_spent', 'amount_spent_y': 'total_amount_spent'})
# df_orders_cust_tiers.head()
```

```python
final_table = final_table.rename(columns = {'amount_spent_x': 'amount_spent', 'amount_spent_y': 'total_spent'})
final_table
```

```python
final_table[final_table['customer_priority']=='VIP']['Country'].mode().iloc[0]
```

```python
display(final_table[(final_table['customer_priority']=='VIP') | (final_table['customer_priority']=='Preferred')]['Country'].mode())
final_table[(final_table['customer_priority']=='VIP') | (final_table['customer_priority']=='Preferred')]['Country'].mode().iloc[0]
```
